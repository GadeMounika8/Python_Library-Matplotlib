# Lines (Style, width, color) in matplotlib|| Line properties in matplotlib:

# Line style:

line style:-
    
    --> it is denoted by ls.
    ls = "solid" --> by default
    ls = "dotted"
    ls = "dashed"
    ls = "dashdot"
    ls = "None"   

# Line width:

-->line width - lw

-->line color - c

ex:-
    plot(x,y, is = "dotted",lw = 20, c = "r")

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,16,2)
y = x**2
print(x,y)
plt.plot(x,y,marker = ".", ms=10, mec="k" ,mfc="m", c="b")
plt.show()
```

# line style, line width & line color:

```python
x = np.arange(0,16,2)
y = x**2
print(x,y)
plt.plot(x,y,marker = "*", ms=10, mec="k" ,mfc="r", c="g",ls = "dotted",lw = 2)
plt.show()
```

```python
x = np.arange(0,16,2)
y = x**3
print(x,y)
plt.plot(x,y,marker = "D", ms=6, mec="k" ,mfc="w", c="r",ls = "dashed",lw = 1.8 )
plt.show()
```

```python
x = np.arange(0,20,2)
y = x**3
print(x,y)
plt.plot(x,y,marker = "o", ms=6, mec="k" ,mfc="y",ls = "dashdot",lw = 2)
plt.show()
```

```python
x = np.arange(0,16,2)
y = x**2
print(x,y)
plt.plot(x,y,ls = "None")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
a = np.arange(0,16,2)
b = a*3
plt.plot(a,b)
plt.title("Multiples of 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Grid Function with Line Properties:

--> Grid function gives the lines on x-axis and y-axis.

--> chart will have grid lines.

--> grid() is used to get gridlines on charts.

grid(axis="x")

grid(axis="y")

grid(axis="both")

--> by default axis = "both"

grid(axis="both",ls='dashdot',color = 'y',lw =10 )

lw - linewidth

c - color

ls - linestyle

```python
x = np.arange(0,20,2)
y = x**3
print(x,y)
plt.plot(x,y,marker = "o", ms=6, mec="k" ,mfc="y",ls = "dashdot",lw = 2)
plt.grid(axis = 'both',ls = 'dotted',lw = 1,c='k')
plt.show()
```
