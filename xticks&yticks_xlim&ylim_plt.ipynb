# xticks(), yticks(), xlim(), ylim():-

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,5,6]
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,5,6]
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,5,6]
plt.plot(x,y)
plt.xticks(x, labels = ['stu_name1', 'stu_name2','stu_name3','stu_name4','stu_name5'])
plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,5,6]
plt.plot(x,y)
plt.xlim(0,9)
plt.ylim(0,10)
plt.show()
```
