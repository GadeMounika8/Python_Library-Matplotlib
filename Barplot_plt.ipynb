# xlim, ylim:-

```python
from matplotlib import pyplot as plt
```

```python
x = [10,25,36,47,53]
y = [20,40,60,80,100]
plt.plot(x,y,marker = "*",mec = "k",ms=10,c = "g")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim([0,80])
plt.ylim([0,120])
plt.show()
```

# xticks, yticks, xlim, ylim:-

```python
x = [10,25,36,47,53]
y = [20,40,60,80,100]
plt.plot(x,y,marker = "*",mec = "k",ls="dashdot",ms = 13)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks(x,[20,40,60,80,100])
plt.yticks(y,[10,30,50,70,90])
plt.xlim([0,100])
plt.ylim([0,120])
plt.show()
```
