# Histogram Plot:

--> Graphical representation of the spread/distribution of supplied data.

--> Also used to see the frequency distribution i.e. to know how often each unique value in the supplied data occurs.

# Syntax:

import matplotlib.pyplot as plt

x = []

plt.hist(x)

plt.show()

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b)
plt.show()
```

# Vertical Histogram Plot:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b)
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# color:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'g')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# edgecolor:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'g', edgecolor = 'k')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# bins:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
bin = [10,20,30,40,50]
plt.hist(b, color = 'g', bins = bin)
plt.xticks([10,20,30,40,50])
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# facecolor:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
bin = [10,20,30,40,50]
plt.hist(b, color = 'g', edgecolor = 'k', bins = bin, facecolor = 'm')
plt.xticks([10,20,30,40,50])
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, 'auto', (0,60), color = 'y', edgecolor = 'k')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# cumulative:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'orange', edgecolor = 'g', cumulative = -1)
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# align:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'purple', edgecolor = 'k', align = 'mid')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'olive', edgecolor = 'k', histtype = 'stepfilled')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'gray', edgecolor = 'k', histtype = 'barstacked')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# Horizontal Histogram Plot:-

# orientation:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'gold', edgecolor = 'k', orientation = 'horizontal')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'cyan', edgecolor = 'k', orientation = 'vertical')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# rwidth:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'm', edgecolor = 'k', rwidth = 0.9)
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# log:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
plt.hist(b, color = 'b', edgecolor = 'k', log = True)
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

# label:

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
bin = [10,20,30,40,50]
plt.hist(b, color = 'm', bins = bin, edgecolor = 'k', rwidth = 0.9, label = 'Label')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
bin = [10,20,30,40,50]
plt.hist(b, color = 'grey', bins = bin, edgecolor = 'k', label = 'Label')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.axvline(35, color = 'r', label = 'line')
plt.legend()
plt.show()
```

```python
b = [25, 18, 33, 47, 46, 38 ,31, 31, 48, 23, 10, 33, 40, 27, 41]
bin = [10,20,30,40,50]
plt.hist(b, color = 'brown', bins = bin, edgecolor = 'k', label = 'Label')
plt.title('Histogram')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.axvline(35, color = 'g', label = 'line')
plt.legend()
plt.grid()
plt.show()
```
