# Box Plot & whisker plot:

--> This provides a summary of supplied data, which includes the information like:
    
        Mean
        Median
        Minimum
        Maximum
        1st Quartile
        3rd Quartile
        Outliers

# Syntax:

import matplotlib.pyplot as plt

x = []

plt.boxplot(x)

plt.show()

# Vertical Box Plot:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x)
plt.show()
```

# notch:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = True)
plt.show()
```

# vert:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = False, vert = False)
plt.show()
```

# widths:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = False, vert = True, widths = 0.3)
plt.show()
```

# labels:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'])
plt.show()
```

# patch_artist:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'], patch_artist = True)
plt.show()
```

# showmeans:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'], patch_artist = False, showmeans = True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9,60]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'], patch_artist = True, showmeans = True)
plt.show()
```

# whis:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9,60]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'], patch_artist = True, showmeans = True, whis = 2)
plt.show()
```

# sym:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9,60]
plt.boxplot(x, notch = False, vert = True, widths = 0.3, labels = ['Students'], patch_artist = True, showmeans = True, sym = "r*")
plt.show()
```

# boxprops, capprops & whiskerprops:

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9,60]
plt.boxplot(x, labels = ['Students'], showmeans = True, sym = "r*", boxprops = dict(color = 'blue'), capprops = dict(color = 'blue'),whiskerprops = dict(color = 'red'))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [15,10,20,30,35,25,32,29,9,60]
plt.boxplot(x, labels = ['Students'], showmeans = True, sym = "r*", boxprops = dict(color = 'blue'), capprops = dict(color = 'blue'), whiskerprops = dict(color = 'red'),
flierprops = dict(markeredgecolor = 'y'))
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [15,10,20,30,35,25,32,29,9,60]
a1 = [10,30,25,20,34,40,46, 41,50,5]
plt.boxplot(a, labels = ['Students'], showmeans = True, sym = "r*", boxprops = dict(color = 'blue'), capprops = dict(color = 'blue'),whiskerprops = dict(color = 'red'))
plt.boxplot(a1, labels = ['Students'], showmeans = True, sym = "r*", boxprops = dict(color = 'blue'), capprops = dict(color = 'blue'),whiskerprops = dict(color = 'red'))
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [15,10,20,30,35,25,32,29,9,60]
a1 = [10,30,25,20,34,40,46, 41,50,5]
b = [a, a1]
plt.boxplot(b, labels = ['Section - I', 'Section - II'], showmeans = True, sym = "r*", boxprops = dict(color = 'blue'), capprops = dict(color = 'blue'),whiskerprops = dict(color = 'red'))
plt.show()
```
