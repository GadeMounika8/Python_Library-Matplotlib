# Legend Function in matplotlib:

--> Legend function is used to describe the functionality of line if you plot multiple lines in a graph.

legend() - without parameters

legend(parameters)

# Parameters:

1. loc = 0 to 10

"upper left"

"upper right"

"lower left"

"lower right"

"center"

"lower center"

"upper center"

2. framealpha = 0.6 or 0 to 1

   framealpha = 0 to 1

3. facecolor = "red" or #457895

4. edge color = "yellow" or hexadecimal

5. shadow = True/False

6. Fancybox = True/False

   plt.legend(loc=,framealpha=,facecolor=,edgecolor=,shadow=,Fancybox=)
    
legend(label= ['squares','cubes']),loc,framealpha,facecolor,edgecolor,shadow,fancybox)

```python
from matplotlib import pyplot as plt
import numpy as np
```

# label:

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares")
plt.plot(a,c,label = "cubes")
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares")
plt.plot(a,c,label = "cubes")
plt.legend(loc = "upper left")
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares")
plt.plot(a,c,label = "cubes")
plt.legend(loc = "upper center")
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares")
plt.plot(a,c,label = "cubes")
plt.legend(loc=2)
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares")
plt.plot(a,c,label = "cubes")
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.show()
```

# label, marker style, marker size & marker edgecolor:

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares", marker = "*", ms = 10, markeredgecolor = 'k')
plt.plot(a,c,label = "cubes", marker = 'o', ms = 8, markeredgecolor = 'k')
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.show()
```

# linewidth, marker facecolor:

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares", marker = "*", ms = 10, markeredgecolor = 'k', linewidth = 2,mfc = 'r')
plt.plot(a,c,label = "cubes", marker = 'o', ms = 8, markeredgecolor = 'k',linewidth = 3, mfc = 'g')
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.show()
```

# linestyle:

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares", marker = "*", ms = 10, markeredgecolor = 'k', linewidth = 2,mfc = 'r', linestyle = 'dashed')
plt.plot(a,c,label = "cubes", marker = 'o', ms = 8, markeredgecolor = 'k',linewidth = 3, mfc = 'g', linestyle = 'dotted')
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares", marker = "*", ms = 10, markeredgecolor = 'k', linewidth = 2,mfc = 'r', linestyle = 'solid')
plt.plot(a,c,label = "cubes", marker = 'o', ms = 8, markeredgecolor = 'k',linewidth = 3, mfc = 'g', linestyle = 'solid')
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.grid(True)
plt.show()
```

```python
a = np.arange(0,16,2)
b = a**2
c = a**3
plt.plot(a,b,label = "squares", marker = "*", ms = 10, markeredgecolor = 'k', linewidth = 1,mfc = 'r', linestyle = 'solid')
plt.plot(a,c,label = "cubes", marker = 'o', ms = 8, markeredgecolor = 'k',linewidth = 2, mfc = 'g', linestyle = 'solid')
plt.legend(loc = "upper left",framealpha = 1,facecolor = "yellow",edgecolor = "black",fancybox = True,shadow = True,fontsize=15)
plt.grid(False)
plt.title('Line Plot')
plt.xlabel('Squares')
plt.ylabel('Cubes')
plt.xlim((0,15))
plt.ylim((0,3000))
plt.show()
```

# Bar Plot(Horizontal and Vertical):

--> Bar chart is a graph that represents the category of data with rectangular bars with length and height.

--> These plots can be plotted in horizontally (or) vertically.

--> We have two functions
  
    1. bar()
    2. barh()
    
--> if we use bar(),it represents bars in vertically.

--> if we use barh(), it represents bars in horizontally.

# Vertical Bar Plot(bydefault):-

```python
from matplotlib import pyplot as plt
a = ['a','b','c','d','e']
b = [20,60,80,40,100]
plt.bar(a,b)
plt.show()
```

Note: By default, width = 0.8

# color, width:

```python
from matplotlib import pyplot as plt
a = ['a','b','c','d','e']
b = [20,60,80,40,100]
plt.bar(a,b,color = 'green',width = 0.4)
plt.show()
```

```python
from matplotlib import pyplot as plt
a = ['a','b','c','d','e']
b = [20,60,80,40,100]
colors = ['green','red','blue','orange','purple']
plt.bar(a,b,width = 0.4,color = colors)
plt.show()
```

```python
from matplotlib import pyplot as plt
a = ['a','b','c','d','e']
b = [20,60,80,40,100]
colors = ['green','red','blue','orange','purple']
plt.bar(a,b,width = 0.4,color = colors)
plt.title('Bar Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Horizontal Bar Plot:-

# height:

```python
from matplotlib import pyplot as plt
a = ['a','b','c','d','e']
b = [20,60,80,40,100]
colors = ['green','red','blue','orange','purple']
plt.barh(a,b,height = 0.4,color = colors)
plt.title('Bar Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
