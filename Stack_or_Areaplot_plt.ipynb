# Stack Plot (or) Area Plot

# Syntax:

import matplotlib.pyplot as plt

x = []

y = []

plt.stackplot(x,y)

plt.show()

```python
import matplotlib.pyplot as plt
a = [5,10,15,20,25,30]
area = [3,9,4,8,6,1]
plt.stackplot(a, area)
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [5,10,15,20,25,30]
area = [3,9,4,8,6,1]
area1 = [2,4,6,8,3,1]
area2 = [1,3,5,7,8,4]
plt.stackplot(a, area, area1, area2)
plt.show()
```

# labels:

```python
import matplotlib.pyplot as plt
a = [5,10,15,20,25,30]
area = [3,9,4,8,6,1]
area1 = [2,4,6,8,3,1]
area2 = [1,3,5,7,8,4]
label = ['area', 'area1', 'area2']
plt.stackplot(a, area, area1, area2, labels = label)
plt.legend()
plt.show()
```

# colors:

```python
import matplotlib.pyplot as plt
a = [5,10,15,20,25,30]
area = [3,9,4,8,6,1]
area1 = [2,4,6,8,3,1]
area2 = [1,3,5,7,8,4]
label = ['area', 'area1', 'area2']
c = ['maroon','gold','g']
plt.stackplot(a, area, area1, area2, labels = label, colors = c)
plt.title('Course')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc = 2)
plt.show()
```

# baseline:

```python
import matplotlib.pyplot as plt
a = [5,10,15,20,25,30]
area = [3,9,4,8,6,1]
area1 = [2,4,6,8,3,1]
area2 = [1,3,5,7,8,4]
label = ['area', 'area1', 'area2']
c = ['maroon','gold','g']
plt.stackplot(a, area, area1, area2, labels = label, colors = c, baseline = 'sym')
plt.title('course')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend(loc = 2)
plt.show()
```
