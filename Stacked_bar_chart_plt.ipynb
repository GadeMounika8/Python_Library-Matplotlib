```python
import matplotlib.pyplot as plt

Day = ['Sun', 'Mon', 'Tue', 'wed', 'Thur', 'Fri', 'Sat']
tip = [9000, 2000, 1500, 1700, 1900, 5000, 7000]

child = [1200, 500, 300, 400, 200, 600, 800]
women = [2000, 700, 800, 300, 400, 7000, 9000]
men = [5800, 5000, 500, 2000, 4000, 6000, 10000]

child_first = [women[i] + men[i] for i in range(len(men))]

plt.title('tips')
plt.bar(Day, child, width = 0.3, color = 'olive')
plt.bar(Day, women, bottom = child, width = 0.3, color = 'r' )
plt.bar(Day, men, bottom = child_first, width = 0.3, color = 'y')

plt.ylim(0,17000)
plt.show()
```
