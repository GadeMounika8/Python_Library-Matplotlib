# Stem Plot

# Syntax:

import matplotlib.pyplot as plt

x = []

y = []

plt.stem(x,y)

plt.show()

# Vertical Stem Plot:-

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x,y)
plt.show()
```

# linefmt:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--")
plt.show()
```

# markerfmt:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--", markerfmt = "r*")
plt.show()
```

# bottom:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--", markerfmt = "rD", bottom = 9)
plt.show()
```

# basefmt, label:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--", markerfmt = "rD", basefmt = 'g', label = 'Label')
plt.legend()
plt.show()
```

# use_line_collection:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--", markerfmt = "rD",bottom = 0, basefmt = 'gold', label = 'Label', use_line_collection = False)
plt.legend()
plt.show()
```

# Horizontal Stem Plot:

# orientation:

```python
import matplotlib.pyplot as plt
x = [10,15,19,20,23,25,30,37]
y = [2,5,6,1,7,9,8,3]
plt.stem(x, y, linefmt = "--", markerfmt = "rD",bottom = 0, basefmt = 'gold', label = 'Label', use_line_collection = False, orientation = 'horizontal')
plt.legend()
plt.show()
```
