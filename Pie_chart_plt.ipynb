```python
import matplotlib.pyplot as plt
Price = [40000,60000,100000, 50000]
Course = ['DA', 'AI', 'DS','FSD']
plt.pie(Price)
plt.show()
```

# labels:

```python
import matplotlib.pyplot as plt
Price = [40000,60000,100000, 50000]
Course = ['DA', 'AI', 'DS','FSD']
plt.pie(Price, labels = Course)
plt.show()
```

# explode:

```python
import matplotlib.pyplot as plt
Price = [40000,60000,100000, 50000]
Course = ['DA', 'AI', 'DS','FSD']
e = [0,0,0.2,0]
plt.pie(Price, labels = Course, explode = e)
plt.show()
```

# autopct:

```python
import matplotlib.pyplot as plt
Price = [40000,60000,100000, 50000]
Course = ['DA', 'AI', 'DS','FSD']
e = [0,0,0.2,0]
c = ['blue', 'y', 'grey', 'red']
plt.pie(Price, labels = Course, explode = e, colors = c, autopct = '%0.1f%%')
plt.show()
```
