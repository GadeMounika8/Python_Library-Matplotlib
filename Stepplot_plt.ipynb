# Step Plot

# Syntax:

import matplotlib.pyplot as plt

x = []

y = []

plt.step(x, y)

plt.show()

```python
import matplotlib.pyplot as plt
a = [2,4,6,8,10]
b = [1,3,5,7,9]
plt.step(a,b)
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [2,5,8,7,3]
plt.step(a,b)
plt.show()
```

# color, marker style, marker size, marker facecolor:

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [2,5,8,7,3]
plt.step(a,b, color = 'g', marker = '*', ms = 13, mfc = 'r')
plt.title('Step Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# label:

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [2,5,8,7,3]
plt.step(a,b, color = 'g', marker = 'o', ms = 8, mfc = 'r', label = 'sales')
plt.title('Step Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid()
plt.show()
```

# marker edgecolor:

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [2,5,8,7,3]
plt.step(a,b, color = 'y', marker = 'o', ms = 8, mfc = 'r', label = 'sales',markeredgecolor = 'g')
plt.title('Step Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc = 2)
plt.show()
```
