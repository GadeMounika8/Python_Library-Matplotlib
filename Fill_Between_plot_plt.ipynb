# Fill_Between Plot 

# Syntax:

import matplotlib.pyplot as plt

x = []

y = []

plt.plot(x,y)

plt.fill_between()

plt.show()

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [1,3,4,5,8]
plt.plot(a,b,color = 'k')
plt.fill_between(a, b)
plt.title('fill_between Plot')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4,5]
b = [1,3,4,5,8]
plt.plot(a,b,color = 'k')
plt.fill_between(x = [2,4], y1 = 2, y2 = 5, color = 'g')
plt.title('fill_between Plot')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([1,2,3,4,5])
b = np.array([1,3,4,5,8])
plt.plot(a,b,color = 'k')
plt.fill_between(a, b, color = 'orange', where = (a >= 2) & (a <=4), alpha = 0.6)
plt.title('fill_between Plot')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```
