```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,5,6,7,8]
plt.suptitle('Different Charts')

plt.subplot(2,2,1)
plt.plot(x,y,color = 'g')
plt.title('Line Chart')

z = [10,15,25,30]

plt.subplot(2,2,2)
plt.pie(z, colors = 'y',)
plt.title('Pie Chart')

m = ['student_1', 'student_2']
n = [10,12,17,25]

plt.subplot(2,2,3)
plt.boxplot(n)
plt.title('Box Plot')

a = [20,30,35,50,60,70]
b = [2,3,5,7,8,9]
plt.subplot(2,2,4)
plt.bar(a,b,color = 'r')
plt.title('Bar Chart')

plt.show()
```
