# Scatter Plot

# Syntax:

```python
import matplotlib.pyplot as plt
x = [10, 15, 20, 17, 25, 20, 23, 29, 30]
y = [5, 3, 9, 8,1, 3,2, 6, 7]
plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
book_price = [1500, 1000, 2000, 1700, 250, 200, 2300, 2900, 3000]
quantity = [5, 3, 9, 8, 1, 3, 2, 6, 7]
plt.scatter(book_price, quantity)
plt.title('Book Details', fontsize = 20)
plt.xlabel('Book_Price', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```

# color:

```python
import matplotlib.pyplot as plt
book_price = [1500, 1000, 2000, 1700, 250, 200, 2300, 2900, 3000]
quantity = [5, 3, 9, 8, 1, 3, 2, 6, 7]
plt.scatter(book_price, quantity, color = 'g')
plt.title('Book Details', fontsize = 20)
plt.xlabel('Book_Price', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```

# colorbar:-

```python
import matplotlib.pyplot as plt
book_price = [1500, 1000, 2000, 1700, 250, 200, 2300, 2900, 3000]
quantity = [5, 3, 9, 8, 1, 3, 2, 6, 7]
col = [15,20,29,35,40,49,52,59,65]
sizes = [130,150,70,100,120,150,170,140,80]
plt.scatter(book_price, quantity, c = col, s = sizes, alpha = 1.0, marker = "*", edgecolor = 'k', linewidth = 1, cmap = 'PiYG')
plt.colorbar()
plt.title('Book Details', fontsize = 20)
plt.xlabel('Book_Price', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```

```python
import matplotlib.pyplot as plt
book_price = [1500, 1000, 2000, 1700, 250, 200, 2300, 2900, 3000]
quantity = [5, 3, 9, 8, 1, 3, 2, 6, 7]
col = [15,20,29,35,40,49,52,59,65]
sizes = [130,150,70,100,120,150,170,140,80]
plt.scatter(book_price, quantity, c = col, s = sizes, alpha = 1.0, marker = "*", edgecolor = 'k', linewidth = 1, cmap = 'viridis')
b = plt.colorbar()
b.set_label('cbar', fontsize = 8)
plt.title('Book Details', fontsize = 20)
plt.xlabel('Book_Price', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```

```python
import matplotlib.pyplot as plt
book_price = [1500, 1000, 2000, 1700, 250, 200, 2300, 2900, 3000]
quantity = [5, 3, 9, 8, 1, 3, 2, 6, 7]
quantity1 = [2,3,6,9,8,6,7,5,4]
col = [15,20,29,35,40,49,52,59,65]
sizes = [130,150,70,100,120,150,170,140,80]
plt.scatter(book_price, quantity, c = col, s = sizes, alpha = 1.0, marker = "*", edgecolor = 'k', linewidth = 1, cmap = 'viridis', label = 'Books')
plt.scatter(book_price, quantity1, c = 'orange', s = sizes, alpha = 1.0, marker = "o", edgecolor = 'k', linewidth = 1)
b = plt.colorbar()
b.set_label('cbar', fontsize = 8)
plt.legend()
plt.title('Book Details', fontsize = 20)
plt.xlabel('Book_Price', fontsize = 12)
plt.ylabel('Quantity', fontsize = 12)
plt.show()
```
